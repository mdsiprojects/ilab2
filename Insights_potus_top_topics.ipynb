```python
import numpy as np
from helpers.topics import get_topic

ntopics = 5
result_model = result.model



iloc = 123

df_weigths_matrix = pd.DataFrame()

for idx, r in result.df.iterrows(): 
    speech_topic_probs = r.probs
    speech_id = r.speech_id

    topics_ordered = np.flip(speech_topic_probs.argsort())

    dft =pd.Series(topics_ordered, name='topic').reset_index().rename(columns={'index': 'weight'})
    dft.weight = np.maximum(0,ntopics-dft.weight)
    dft['speech_id'] = speech_id
    
    df_weigths_matrix = df_weigths_matrix.append(dft)
```

```python
top_topics_by_weight = df_weigths_matrix.groupby('topic').sum('weight').sort_values('weight', ascending=False)
display(top_topics_by_weight)

sum(top_topics_by_weight.weight)
```
